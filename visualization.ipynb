```python
key_path = "../../Downloads/buring-case-credentials.json"
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
credentials = service_account.Credentials.from_service_account_file(key_path)
client = bigquery.Client(credentials=credentials, project=credentials.project_id)
```

```python
import altair as alt
```

```python
q1 = """
SELECT repo_name, usage_count
FROM `bruin-hiring.cdt_caneraslan.q1`
ORDER BY usage_count DESC
"""

res = client.query(q1).result()
df = res.to_dataframe()

alt.Chart(df).mark_bar().encode(
    x=alt.X('repo_name:N', title='Repository Name'),
    y=alt.Y('usage_count:Q', title='Usage Count')
)
```

```python
q2 = """
SELECT *
FROM `bruin-hiring.cdt_caneraslan.q2`
ORDER BY contribution_number DESC
LIMIT 10
"""

query_job = client.query(q2)
res = query_job.result()

df = res.to_dataframe()
# print(df)

alt.Chart(df).mark_bar().encode(
    x=alt.X('contributor:N', title='Contributor Name'),
    y=alt.Y('contribution_number:Q', title='Contribution Number')
)
```

```python
q5 = """
SELECT *
FROM bruin-hiring.cdt_caneraslan.q5
ORDER BY number_of_commits DESC
LIMIT 10
"""

res = client.query(q5).result()
df = res.to_dataframe()

df['commit_date'] = pd.to_datetime(df['commit_date']) # we cannot directly convert dbtype to string
df['commit_date'] = df['commit_date'].dt.strftime('%Y-%m-%d')

alt.Chart(df).mark_bar().encode(
    y=alt.Y('commit_date:N', title='Commit Date'),
    x=alt.X('number_of_commits:Q', title='Number of Commits')
)
```
