```python
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
import altair as alt

key_path = "../../Downloads/bruin-case-credentials.json"

credentials = service_account.Credentials.from_service_account_file(key_path)
client = bigquery.Client(credentials=credentials, project=credentials.project_id)
```

```python
# 1- What are the most popular repos?

q1 = """
SELECT *
FROM `bruin-hiring.cdt_caneraslan.q1`
ORDER BY 2 DESC
"""

res = client.query(q1).result()
df = res.to_dataframe()

alt.Chart(df).mark_bar().encode(
    x=alt.X('repo_name:N', title='Repository Name', sort='-y'),
    y=alt.Y('commit_number:Q', title='Commit Count')
)
```

```python
# 2- Who are the top contributors?

q2 = """
SELECT *
FROM `bruin-hiring.cdt_caneraslan.q2`
ORDER BY contribution_number DESC
LIMIT 10
"""

query_job = client.query(q2)
res = query_job.result()

df = res.to_dataframe()
# df

# how to disable aggregation???

alt.Chart(df, width=400).mark_bar().encode(
  x=alt.X('contributor:N', title='Contributor Name'),
  y=alt.Y('contribution_number:Q', title='Contribution Number', aggregate='none') 
)
```

```python
# 3- Are there any developer contributing to more than 1 repos?

q3 = """
SELECT *
FROM `bruin-hiring.cdt_caneraslan.q3`
ORDER BY 2 DESC
LIMIT 10
"""

res = client.query(q3).result()
df = res.to_dataframe()
# print(res.to_dataframe())

# kac kisi 1,2,3.. repoya gitti?

alt.Chart(df).mark_bar().encode(
    x=alt.X('contributor:N', title='Contributor Name', sort='-y'),
    y=alt.Y('distinct_repo_number:Q', title='Distinct Repo Number')
)
```

```python
# 4- What are the most popular email domains?

q4 = """
SELECT *
FROM `bruin-hiring.cdt_caneraslan.q4`
ORDER BY commit_number DESC
LIMIT 25
"""

res = client.query(q4).result()
df = res.to_dataframe()
# print(df)

alt.Chart(df).mark_bar().encode(
    y=alt.Y('email_domain:N', title='Email Domain'),
    x=alt.X('commit_number:Q', title='Commit Number')
)
```

```python
# 5- Number of commits per day

q5 = """
SELECT *
FROM bruin-hiring.cdt_caneraslan.q5
ORDER BY commit_number DESC
"""

res = client.query(q5).result()
df = res.to_dataframe()

df['commit_date'] = pd.to_datetime(df['commit_date']) # we cannot directly convert dbtype to string

alt.Chart(df, width=800).mark_bar().encode(
    x=alt.X('commit_date', title='Commit Date'),
    y=alt.Y('commit_number:Q', title='Commit Number'),
    tooltip=['commit_date']
)
```

```python
# 6- Number of commits per weekday

q6 = """
SELECT 
    EXTRACT(DAYOFWEEK FROM committer.date) AS commit_date,
    COUNT(committer.date) AS number_of_commits
FROM bruin-hiring.github_repos.sample_commits
GROUP BY 1
ORDER BY 1 ASC
"""
# print(df)
res = client.query(q6).result()
df = res.to_dataframe()
df['pct'] = df['number_of_commits']/df['number_of_commits'].sum()

alt.Chart(df, width=800).mark_bar().encode(
    x=alt.X('commit_date:N', title='Commit Date'),
    y=alt.Y('pct:Q', title='pct').axis(format='.0%'),
)
```
